```python
import pandas as pd,numpy as np

df = pd.read_csv("D:\\Data_Science\\Regression\\Advertising.csv")

```

```python
df['Theta_One'] = np.ones((200))
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['TV'] = scaler.fit_transform(df[['TV']])
```

```python
x=df[['TV','Theta_One']]
y=df['Sales']

x=np.array(x)
y=np.array(y)
```

```python
m,n=x.shape
theta = np.zeros((n))
```

```python
m
```

- x                                       = m*n
- x_transpose                             = n*m
- theta                                   = n*1      
- hypothesis = x.theta = (m*n).(n*1)      = m*1
- y                                       = m*1
- loss                                    = m*1
- gradient = x_trans.loss = (n*m).(m*1)   = n*1
- theta and gradient are of same shape
- dot product is same as taking derivative.

```python
alpha=0.005
xTrans = x.transpose()
for i in range(10000):
    hypothesis = np.dot(x, theta)             ## Prediction
    loss = hypothesis - y                     ## Loss function
    cost = np.sum(loss ** 2) / m              ## Cost Function = average of loss functions
    print("Iteration %d | Cost: %f" % (i, cost))
    # avg gradient per example
    gradient = np.dot(xTrans, loss) / m       ## Gradient
    # update
    theta = theta - alpha * gradient          ## Alpha determines the size of step towards the gradient.
```

```python
y_pred=np.dot(x,theta)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```
