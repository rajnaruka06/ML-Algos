```python
import pandas as pd,numpy as np
import seaborn as sns
df = pd.read_csv("D:\\Data_Science\\Regression\\Advertising.csv")

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x=df.drop(columns='Sales')
x = scaler.fit_transform(x)
y=df['Sales']
```

```python
df.columns
```

```python
df1 = pd.DataFrame(x,columns=['TV', 'Radio', 'Newspaper'])
df1['Theta_One'] = np.ones((200))
```

```python
x=df1.copy()
```

```python
x=np.array(x)
y=np.array(y)
```

```python
m,n=x.shape
theta = np.zeros((n))
```

```python
alpha=0.01
xTrans = x.transpose()
for i in range(10000):
    hypothesis = np.dot(x, theta)
    loss = hypothesis - y
    cost = np.sum(loss ** 2) / (2*m)
    #print(gradient)
    print("Iteration %d | Cost: %f" % (i, cost))
    # avg gradient per example
    gradient = np.dot(xTrans, loss) / (2*m)
    # update
    theta = theta - alpha * gradient
```

```python
y_pred = np.dot(x,theta)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```
